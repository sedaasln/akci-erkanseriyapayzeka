```python
img_list = []

from tqdm import tqdm_notebook as tqdm

for i in tqdm(files):
    image = cv2.imread(path + 'adenocarcinoma\\'+i)
    image = cv2.resize(image,(400,400))
    image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
    kopya = image.copy()
    kopya = cv2.cvtColor(kopya, cv2.COLOR_RGB2GRAY)
    blur = cv2.GaussianBlur(kopya,(5,5),0)
    thresh = cv2.threshold(blur,10,255, cv2.THRESH_BINARY)[1]
    lab = cv2.cvtColor(image, cv2.COLOR_RGB2LAB)
    l,a,b = cv2.split(lab)
    clahe = cv2.createCLAHE(clipLimit=5.0,tileGridSize=((8,8)))
    cl = clahe.apply(l)
    limg = cv2.merge((cl,a,b))
    son = cv2.cvtColor(limg, cv2.COLOR_LAB2RGB)
    med_son = cv2.medianBlur(son, 3)
    arka_plan = cv2.medianBlur(son, 37)
    maske = cv2.addWeighted(med_son,1,arka_plan,-1,255)
    son_img = cv2.bitwise_and(maske,med_son)
    img_list.append(son_img)
```

```python
print("\n\n\t\tTraining Set")
print("\t  ========================\n")
train_c = []
for folder in os.listdir(train_path):
    train_c.append(folder)
    print("\nTrain "+ folder + " Class: ", len(os.listdir(train_path + '/' + folder)))

print("\n\n", train_c)

print("\n\n\t\tTesting Set")
print("\t  ========================\n")
test_c = []
for folder in os.listdir(test_path):
    test_c.append(folder)
    print("\nTest "+ folder + " Class: ", len(os.listdir(test_path + '/' + folder)))

print("\n\n", test_c)

print("\n\n\t\tValidation Set")
print("\t  ========================\n")
val_c = []
for folder in os.listdir(valid_path):
    val_c.append(folder)
    print("\nValidation "+ folder + " Class: ", len(os.listdir(valid_path + '/' + folder)))

print("\n\n", val_c, "\n")
```

```python
image_shape = (400,400,3)
N_CLASSES = 4


train_datagen = ImageDataGenerator(dtype='float32', preprocessing_function=densenet.preprocess_input)
train_generator = train_datagen.flow_from_directory(train_path,
                                                   batch_size =32,
                                                   target_size = (400,400),
                                                   class_mode = 'categorical')

valid_datagen = ImageDataGenerator(dtype='float32', preprocessing_function=densenet.preprocess_input)
valid_generator = valid_datagen.flow_from_directory(valid_path,
                                                   batch_size =32,
                                                   target_size = (400,400),
                                                   class_mode = 'categorical')

test_datagen = ImageDataGenerator(dtype='float32', preprocessing_function=densenet.preprocess_input)
test_generator = test_datagen.flow_from_directory(test_path,
                                                   batch_size = 32,
                                                   target_size = (400,400),
                                                   class_mode = 'categorical')
```

```python
dense_model = DenseNet201(include_top=False, pooling='avg', weights='imagenet', input_shape = (image_shape))

# conv5 katmanlar dndaki tm katmanlar eitilemez hale getiriyoruz
for layer in dense_model.layers:
    if 'conv5' not in layer.name:
        layer.trainable = False
```

```python
densenet_model = Sequential()
densenet_model.add(dense_model)
densenet_model.add(Flatten())
densenet_model.add(BatchNormalization())
densenet_model.add(Dense(N_CLASSES, activation='softmax')) # aktivasyon fonksiyonunu belirliyoruz
densenet_model.summary()
```

```python
optimizer = optimizers.Adam(learning_rate= 0.00001, weight_decay= 1e-6)# renme katsaysn ve arl belirliyoruz
densenet_model.compile(optimizer=optimizer, loss = 'categorical_crossentropy', metrics = ['acc'],loss_weights=0.001)

checkpointer = ModelCheckpoint(filepath='./chestmodel-DenseNet201.hdf5',
                            monitor='val_loss', verbose = 1,
                            save_best_only=True)
early_stopping = EarlyStopping(verbose=1, patience=15)
```

```python
history_dense = densenet_model.fit(train_generator,
                                steps_per_epoch = 20,
                                epochs = 10,
                                verbose = 1,
                                validation_data = valid_generator,
                                callbacks = [checkpointer, early_stopping])
```

```python
plt.plot(history_dense.history['acc'], label = 'train',)
plt.plot(history_dense.history['val_acc'], label = 'valid')

plt.legend(loc = 'upper left')
plt.xlabel('epochs')
plt.ylabel('accuracy')
plt.show()
```

```python
dns_result = densenet_model.evaluate(test_generator)
```
